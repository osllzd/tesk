```python
# 이상치 계산하기
def print_filer(df):
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)
    iqr = q3 - q1
    print(f'[q1]\n{q1}\n\n[q3\n{q3}\n')
    print(f'[iqr의 범위]\n{iqr}\n')
    print('이상값이 될 기준 계산\n')
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    mask = df < lower
    print(f'lower의 개수:{mask.sum()}')
    mask2 = df > upper
    print(f'upper의 개수:{mask2.sum()}')
    return lower, upper
```

```python
lower, upper = print_filer(houDF)
```

```python
import pandas as pd

# DataFrame 생성 예시
data = houDF
df = pd.DataFrame(data)

# 이상치 식별 함수 정의
def identify_outliers(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    return ((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR)))

# 전체 데이터프레임에서 이상치 식별
outliers = identify_outliers(df)

# 이상치 제거
for column in df.columns:
    df = df[~outliers[column]]

print(df)
```

```python
from sklearn.linear_model import Lasso
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

# 데이터 로드
X, y = featureDF, targetSR

# 데이터 스케일링
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=500)

# Lasso
lasso = Lasso(alpha=0.1)  # alpha 값은 규제의 강도를 조절
lasso.fit(X_train, y_train)

# 모델 평가
train_score = lasso.score(X_train, y_train)
test_score = lasso.score(X_test, y_test)
print("Training R2 score:", train_score)
print("Test R2 score:", test_score)

# 예측
y_pred = lasso.predict(X_test)

# 평가 지표 출력
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

```

```python
# Ridge
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=1.0)  # alpha 값은 규제의 강도를 조절합니다.
ridge.fit(X_train, y_train)

# 모델 평가
train_score = ridge.score(X_train, y_train)
test_score = ridge.score(X_test, y_test)
print("Training R2 score:", train_score)
print("Test R2 score:", test_score)


y_pred = ridge.predict(X_test)


mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```
